Note, however, that the new data frame cases_df is simply a "view" of the original data frame covid_df i.e. they both point to the same data in the computer's memory, and changing any values inside one of them will also change the respective values in the other. Sharing data between data frames makes data manipulation in Pandas blazing fast, and you needn't worry about the overhead of copying thousands or millions of rows every time you want to crate a new data frame by operating on an existing one.

Sometimes you might need a full copy of the data frame, in which case you can use the copy method.

```python
covid_df_copy = covid_df.copy()
```

```python
covid_df_copy
```

The data within covid_df_copy is completely separate from covid_df ,and changing values inside one of them will be not affect the other .

## Anylyzing Data from data frame

Q . What is the total number of reported cases and deaths related to covid-19 in Italy ? 

Similar to NumPy arrays , Pandas series supports the sum  method to answer these questions.

```python
total_cases = covid_df.new_cases.sum()
total_deaths = covid_df.new_deaths.sum()
```

```python
print(f'The number of reported cases in {total_cases} and the number of deaths is {total_deaths} ')
```

Q : What is THe overall deaths rate (ratio of reported deats to reported cases)?


```python
death_rate = covid_df.new_deaths.sum()/covid_df.new_cases.sum()
```

```python
print("THe overall reported death rate in italy is {:.2f}%.".format(death_rate))
```

Q : What is the overall number of tests conducted ? A total of 935310 tests were conducted before daily test numbers were being reported .


we can check the first non-NaN index using first_valid_index

```python
initial_tests = 935310

total_tests = initial_tests + covid_df.new_tests.sum()
```

```python
total_tests
```

Q : What fraction of test returned a positive result ?
 

```python
positive_rate = total_cases / total_tests
```

```python
print('{:.2f}%  of tests in Italy led to a positive diagnosis.'.format(positive_rate))
```
