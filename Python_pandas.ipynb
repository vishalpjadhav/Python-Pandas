## Pandas is a popular open-source Python library used for data analysis and data manipulation. It’s especially useful when working with structured data like tables

# What Pandas is used for
Cleaning messy data (missing values, duplicates, formatting)
Analyzing datasets (statistics, filtering, grouping)
Reading/writing data (CSV, Excel, SQL, JSON, etc.)
Time series analysis (dates, trends)
Data transformation (merging, reshaping)

## Core concepts
* Series → a single column of data (like a list with labels)
* DataFrame → a table of rows and columns (most commonly used)
